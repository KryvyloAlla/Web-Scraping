```python
import requests
from bs4 import BeautifulSoup
import json

url = 'https://www.binance.com/ru-UA'

response = requests.get(url)
html_content = response.content

soup = BeautifulSoup(html_content, 'html.parser')

crypto_data = []

cryptos = [
    {'name': 'Bitcoin', 'ticker': 'BTC', 'IDS': 'top_crypto_table-1-BTC_USDT'},
    {'name': 'Ethereum', 'ticker': 'ETH', 'IDS': 'top_crypto_table-2-ETH_USDT'}   
]

for crypto in cryptos:
    name = crypto['name']
    ticker = crypto['ticker']
    ids = crypto['IDS']
    # Знаходимо назву криптовалюти
    name_tag_a = soup.find('a', {"id": ids})
    print(name_tag_a)
    name_tag = name_tag_a.find_all("div", {"class": "whitespace-nowrap text-body2 text-t-third"})
    print(name_tag)
    
    # Знаходимо курс криптовалюти
    if name_tag:
        price_tag = name_tag_a.find('div', class_='lock-ltr text-subtitle1 font-medium')
        if price_tag:
            price = price_tag.text
            crypto_data.append({'name': name, 'ticker': ticker, 'price': price})
            print(price)

with open('crypto_data_Binance.json', 'w') as f:
    json.dump(crypto_data, f, ensure_ascii=False, indent=4)

print("Дані збережено у файл crypto_data_Binance.json")
```

```python
import requests
from bs4 import BeautifulSoup
import json
import requests
from bs4 import BeautifulSoup
import json

url = 'https://www.gate.io/ru'

response = requests.get(url)
html_content = response.content

soup = BeautifulSoup(html_content, 'html.parser')

crypto_data = []

cryptos = [
    {'name': 'Bitcoin', 'ticker': 'BTC', 'IDS': 1},
    {'name': 'Ethereum', 'ticker': 'ETH', 'IDS': 0}
    
]

for crypto in cryptos:
    name = crypto['name']
    ticker = crypto['ticker']
    ids = crypto['IDS']
    # Знаходимо назву криптовалюти
    name_tag_a = soup.find_all('tr', {"class": "mantine-ivepx5 mantine-Table-dataRow"})
   # print(name_tag_a)

    name_tag_b = name_tag_a[ids]
    name_tag = name_tag_b.find('a', {"title": ticker})
    print(ticker)
    
    # Знаходимо курс криптовалюти
    if name_tag:
        price_tag_b = name_tag_b.find_all('td', class_='mantine-10v3ran mantine-Table-dataItem')
        price_tag = price_tag_b[1]
        if price_tag:
            price = price_tag.text
            crypto_data.append({'name': name, 'ticker': ticker, 'price': price})
            print(price)

with open('crypto_data_Gate.json', 'w') as f:
    json.dump(crypto_data, f, ensure_ascii=False, indent=4)

print("Дані збережено у файл crypto_data_Gate.json")
```

```python
import requests
from bs4 import BeautifulSoup
import json
import requests
from bs4 import BeautifulSoup
import json

url = 'https://minfin.com.ua/ua/crypto/articles/'

response = requests.get(url)
html_content = response.content

soup = BeautifulSoup(html_content, 'html.parser')

crypto_data = []

cryptos = [
    {'IDS': 0},
    {'IDS': 1},
    {'IDS': 2}   
]

for crypto in cryptos:

    ids = crypto['IDS']
    # Знаходимо заголовки новини
    name_tag_a = soup.find('div', {"class": "js-news news default-padding"})
    #print(name_tag_a)

    name_tag_b = name_tag_a.find_all('a')[ids]
    name_tag = name_tag_b.text
    print(name_tag)
    
    # Отримуємо link на новини
    link = "https://minfin.com.ua/ua"+name_tag_b.get('href')
    print(link)
    crypto_data.append({'name': name_tag, 'link': link})

with open('crypto_data_News.json', 'w', encoding='utf-8') as f:
    json.dump(crypto_data, f, ensure_ascii=False, indent=4)

print("Дані збережено у файл crypto_data_News.json")
```
